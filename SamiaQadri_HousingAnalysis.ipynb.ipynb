```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df= pd.read_csv('Housing.csv')
```

```
print("Studying the Data: ")
print("Starting five rows: ")
print(df.head())
```

```
print("")
print("\n Observations taken from whole data : ")
print(df.describe(include='all'))
```

```
print("Checking Missing Values")
print(df.isnull().sum())
# now it will remove such value if any
df=df.dropna()
```

```
targetted_columns=['price','area','bedrooms']
for col in targetted_columns:
  print("Data for ",col)
  print("mean      : ", df[col].mean())
  print("Median    :", df[col].median())
  print("Mode      :", df[col].mode()[0])
  print("Variance  :", df[col].var())
  print("Std Dev   :", df[col].std())
  print("Skewness  :", df[col].skew())
  print("Kurtosis  :", df[col].kurt())
  print("Min       :", df[col].min())
  print("Max       :", df[col].max())
  print("")
```

```
# graph poltting for visualization:
# Histogram for price vs area
df['price'].hist(bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

df['area'].hist(bins=30)
plt.title("Area Distribution")
plt.xlabel("Area (sq ft)")
plt.ylabel("Frequency")
plt.show()
```

```
# box-plot graph
sns.boxplot(x=df['price'])
plt.title ("PRICE BOX PLOT ")
plt.show()

sns.boxplot(x=df['area'])
plt.title("AREA BOX PLOT ")
plt.show()
```

```
# Scatter Plot to view trend of price vs area
sns.scatterplot(x='area',y='price',data=df)
plt.title( "PRICE VS AREA ")
plt.show()
```
